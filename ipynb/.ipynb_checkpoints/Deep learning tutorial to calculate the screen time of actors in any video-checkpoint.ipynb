```python
import tensorflow as tf
from skimage.transform import resize
import cv2
import math
import matplotlib.pyplot as plt
import pandas as pd
import os
from keras.preprocessing import image
import numpy as np
from keras.utils import np_utils
```

## Step 1: Read the video, extract frames from it and save them as images

```python
Data_folder = '/Volumes/McVeighLab/projects/Zhennong/Video_Prediction/Synthesized_EF_movie/movie/'
videoFile = os.path.join(Data_folder,'ucsd_toshiba_023_70.7%.avi')
cap = cv2.VideoCapture(videoFile)
frameRate = 1

count = 0
x = 1
while(cap.isOpened()):
    frameId = cap.get(1) # current frame number
    ret, frame = cap.read()
    
    if (ret != True):
        break
    if (frameId % math.floor(frameRate) == 0):
        filename = "frame%d.jpg" % count
        count += 1
        frame_save_path = os.path.join(Data_folder,filename)
        cv2.imwrite(frame_save_path,frame)
cap.release()
print('Done!')
```

## Step 2: label a few images for training the model

```python
# import the label
data = pd.read_csv(os.path.join(Data_folder,'Label for each frame.csv'))
```

```python
X = []
for img_name in data.Image_ID:
    img = plt.imread(os.path.join(Data_folder,img_name))
    X.append(img)

X = np.asarray(X)

# reshape image to fit the VGG neural network (224x224x3)
image = []
for i in range(0,X.shape[0]):
    a = resize(X[i],preserve_range = True, output_shape = (224,224)).astype(int)
    image.append(a)
X = np.asarray(image)
print(X.shape,X[0].shape)
```

```python
y = data.Class
one_hot_y = np_utils.to_categorical(y)
```

```python
# preprocessing the input for VGG
from keras.applications.vgg16 import preprocess_input
X = preprocess_input(X,mode = 'tf')
```

```python
# prepare the training and validation
from sklearn.model_selection import train_test_split
X_train,X_valid,y_train,y_valid = train_test_split(X,one_hot_y,test_size = 0.3,random_state = 42)
print(y_train.shape,y_valid.shape)
```

## step 3: build the model
We will make predictions use this model for X_train and X_valid, get the geatures and then use those features to retrain the model

```python
# first let's look at pre-trained VGG
from keras.applications.vgg16 import VGG16
pre_model = VGG16()
print(pre_model.summary())
```

```python
X_train_VGG_trial = pre_model.predict(X_train[6:7])
X_valid_VGG_trial = pre_model.predict(X_valid[6:7])
print(X_valid_VGG_trial.shape)

# try to denode the prediction
from keras.applications.vgg16 import decode_predictions
# convert the probabilities to class labels
label = decode_predictions(X_valid_VGG_trial)
# retrieve the most likely result, e.g. highest probability
label = label[0][0]
# print the classification
print('%s (%.2f%%)' % (label[1], label[2]*100))
```

```python
# now let's apply VGG to our data
# Here we remove the top(fully-connected layer + softmax), so the model ends with last convolutional layer
base_model = VGG16(weights='imagenet',include_top=False,input_shape=(224,224,3))
X_train_VGG = base_model.predict(X_train)
X_valid_VGG = base_model.predict(X_valid)
print(X_valid_VGG.shape)
```

```python
# now we extract the features from VGG, the next step is to feed features into neural network
# first we need to reshape it to 1-D
X_train_VGG = X_train_VGG.reshape(X_train.shape[0],7*7*512)
X_valid_VGG = X_valid_VGG.reshape(X_valid.shape[0],7*7*512)
print(X_valid_VGG.shape)

# second we preprocess the images and make them zero-centered which help the model to converge faster
X_train_VGG_centered = X_train_VGG/X_train_VGG.max()
X_valid_VGG_centered = X_valid_VGG/X_train_VGG.max()
```

```python
# build our own neural network
from keras.models import Sequential
from keras.layers import Dense, InputLayer, Dropout
# 1. build model
model = Sequential()
model.add(InputLayer((7*7*512,)))    # input layer
model.add(Dense(units=1024, activation='sigmoid'))   # hidden layer
model.add(Dropout(0.5))      # adding dropout, so there is sync on training and validation
model.add(Dense(units=512, activation='sigmoid'))    # hidden layer, increase number of layers to increase accuracy on training
model.add(Dropout(0.5))      # adding dropout
model.add(Dense(units=256, activation='sigmoid'))    # hidden layer
model.add(Dropout(0.5))      # adding dropout
model.add(Dense(3, activation='softmax'))            # output layer

# 2. compile the model
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])

# 3. train
# there is inbalance of Tom and Jerry in the video, so we use class_weights to assign high weight to low value count
from sklearn.utils.class_weight import compute_class_weight, compute_sample_weight
class_weights = compute_class_weight('balanced',np.unique(data.Class), data.Class)  # computing weights of different classes

# add callback
from keras.callbacks import ModelCheckpoint
filepath=os.path.join(Data_folder,'weights_best_for_Tom and Jerry.hdf5')
checkpoint = ModelCheckpoint(filepath, monitor='val_loss', verbose=1, save_best_only=True, mode='min')
callbacks_list = [checkpoint]      # model check pointing based on validation loss

model.fit(X_train_VGG_centered,y_train,epochs = 80,validation_data=(X_valid_VGG_centered,y_valid),class_weight=class_weights,callbacks=callbacks_list)
```

## step 4: calculating the screen time - a simple solution, test our model

```python
videoFile = os.path.join(Data_folder,'Tom and Jerry 3.mp4')
cap = cv2.VideoCapture(videoFile)
frameRate = cap.get(5)

count = 0
while(cap.isOpened()):
    frameId = cap.get(1)
    ret,frame = cap.read()
    if (ret!=True):
        break
    if (frameId % math.floor(frameRate) == 0):
        filename = 'test%d.jpg' % count 
        file_save_path = os.path.join(Data_folder,'test%d.jpg' % count)
        count += 1
        cv2.imwrite(file_save_path,frame)

cap.release()
print('Done!')
```

```python
test_set = pd.read_csv(os.path.join(Data_folder,'testing.csv'))
test_image = []
for img_name in test_set.Image_ID:
    img = plt.imread(os.path.join(Data_folder,img_name))
    test_image.append(img)
test_img = np.asarray(test_image)
# reshape
test_img_reshape = []
for i in range(0,test_img.shape[0]):
    a = resize(test_img[i],preserve_range=True,output_shape = (224,224)).astype(int)
    test_img_reshape.append(a)
test_img_reshape = np.asarray(test_img_reshape)
print(test_img_reshape.shape)
```

```python
test_image = preprocess_input(test_img_reshape,mode='tf')
test_image = base_model.predict(test_image)
test_image = test_image.reshape(test_img_reshape.shape[0],7*7*512)
test_image = test_image / test_image.max()
```

```python
test_y = np_utils.to_categorical(test_set.Class)
```

```python
model.load_weights(os.path.join(Data_folder,'weights_best_for_Tom and Jerry.hdf5'))
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
scores = model.evaluate(test_image,test_y)
print('acc: %.2f%%' %(scores[1]*100))
```

```python
# or
predict = model.predict(test_image)
P = []
for i in predict:
    if np.where(i == np.amax(i))[0] == 0:
        P.append(0)
    elif np.where(i == np.amax(i))[0] == 1:
        P.append(1)
    else:
        P.append(2)
        
GT = []
for i in test_y:
    if np.where(i == np.amax(i))[0] == 0:
        GT.append(0)
    elif np.where(i == np.amax(i))[0] == 1:
        GT.append(1)
    else:
        GT.append(2)

# compare P with test_y
count = 0
for j in range(0,len(P)):
    if P[j] == GT[j]:
        count = count +1
acc = count / len(P)
print(acc)
```

```python
predictions = model.predict_classes(test_image)
print("The screen time of JERRY is", predictions[predictions==1].shape[0], "seconds")
print("The screen time of TOM is", predictions[predictions==2].shape[0], "seconds")
```
